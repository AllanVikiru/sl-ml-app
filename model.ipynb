### Import Dataset and Libs

```python
import pandas as pd
```

```python
df = pd.read_csv('Groceries_dataset.csv')
df.head()
```

### Data Preparation

```python
# prepare transaction IDs
df['transactionId'] = df['Member_number'].astype(str)+'_'+df['Date'].astype(str)
df = df.drop('Member_number', axis='columns')
df = df.drop('Date', axis='columns')
df.head()
```

#### Generate single transactions 

```python
groceries_single = pd.crosstab(df['transactionId'], df['itemDescription'])
groceries_single.head()
```

```python
# generate single transactions
def encode(item_freq):
    res = 0
    if item_freq > 0:
        res = 1
    return res
    
basket_input = groceries_single.applymap(encode)
basket_input
```

```python
# apriori
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

frequent_itemsets = apriori(basket_input, min_support=0.001, use_colnames=True)

rules = association_rules(frequent_itemsets, metric="lift")

rules.head()
```

```python
# top 10 most bought items
rules.sort_values(["support", "confidence","lift"],axis = 0, ascending = False).head(10)
```
